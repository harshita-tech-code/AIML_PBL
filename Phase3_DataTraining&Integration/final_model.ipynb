```python
# Setup and Imports
import pandas as pd
import joblib # Useful for saving trained models
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
# Train the Final Chosen Model
df = pd.read_csv("../Phase1_DataCollection/data/data_train.csv", sep=";")
df = df.drop("id", axis=1)
df["age"] = df["age"] / 365
```

```python
X = df.drop("cardio", axis=1)
y = df["cardio"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

final_model = LogisticRegression()
final_model.fit(X_scaled, y)
print("Final Model trained and ready.")
```

```python
# Integration Function (The "Predictor")
def predict_cardio(input_data):
    """
    Takes a dictionary of patient data, cleans it, and returns a prediction.
    """
    # Convert input to DataFrame
    input_df = pd.DataFrame([input_data])
    
    input_df["age"] = input_df["age"] / 365
    
    input_scaled = scaler.transform(input_df)
    
    # Predict
    prediction = final_model.predict(input_scaled)
    return "Risk Detected" if prediction[0] == 1 else "No Risk"
```

```python
# Testing with a Dummy Patient
new_patient = {
    'age': 18000, 'gender': 2, 'height': 170, 'weight': 80,
    'ap_hi': 120, 'ap_lo': 80, 'cholesterol': 1, 'gluc': 1,
    'smoke': 0, 'alco': 0, 'active': 1
}
```

```python
result = predict_cardio(new_patient)
print(f"Prediction for new patient: {result}")
```
